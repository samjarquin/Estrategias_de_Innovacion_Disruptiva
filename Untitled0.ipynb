```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = px.box(tablita,
             title="📊 Distribución de Calificaciones por Estudiante",
             labels={"value": "Calificación", "variable": "Estudiante"})

# Personalizar el gráfico
fig.update_layout(
    template="plotly_white",
    font=dict(size=12),
    hovermode="x unified",
    height=600,
    width=800
)

fig.update_traces(boxpoints='all', jitter=0.3, pointpos=-1.8)  # Mostrar todos los puntos

# Mostrar el gráfico
fig.show()
```
